**Method to link rooms together**

STEP 17:  Next, you need to add a method so that you can link rooms together. Add this new method below the other methods.
This method takes three parameters : the object itself, the room object to link to, and the relative direction of this object.


```python
def link_room(self, room_to_link, direction):
    self.linked_rooms[direction] = room_to_link
```
