# What is Procedural Programming?
Programming languages that use the procedural paradigm are made up of a series of procedures intended to be run one after another.​

These procedures are made up of statements including variable assignment, selection (IF), iteration (loops) and procedures.​

Modularity (each part of the code is self-contained and reusable) is introduced through procedures, from which the paradigm gets its name.

# Understanding the __init__() Function
**Purpose:** The __init__() function initializes an object's state by setting the initial values for its attributes. It is automatically called when a new instance of the class is created.

**Syntax:** The method name is __init__, with double underscores before and after "init". It typically takes at least one parameter, self, which refers to the instance being created.

**Parameters:** Besides self, the __init__() function can take any number of additional parameters required to initialize the object.

```python
class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade
```

*  STEP 4: Give your new object some data in the relevant parameters

```python
# Creating an instance of Student
student1 = Student("Alice", 20, "A")

# Accessing attributes
print(student1.name)  # Output: Alice
print(student1.age)   # Output: 20
print(student1.grade) # Output: A
```

# Understanding the __str__() Function
**Purpose:**
* Readable Representation: The primary purpose of the __str__() method is to return a readable, human-friendly string representation of an object.

* Printing Objects: When you print an object, the __str__() method is automatically called to determine what gets printed.

* Conversion to String: When you pass an object to the str() function, it calls the __str__() method to get a string representation.

**Syntax:** The __str__() method is defined within a class and takes only one parameter: self, which refers to the instance of the class.


class ClassName:

    def __str__(self):

        return "string representation"

```python
class Student:
    def __init__(self, name, age, grade):
        self.name = name  # Initialize the 'name' attribute
        self.age = age    # Initialize the 'age' attribute
        self.grade = grade  # Initialize the 'grade' attribute

    def __str__(self):
        return f"Student Name: {self.name}, Age: {self.age}, Grade: {self.grade}"

    def update_grade(self, new_grade):
        self.grade = new_grade  # Modifies the instance attribute 'grade'

# Example Usage
student1 = Student("Alice", 20, "A")

# Accessing attributes
print(student1.name)  # Output: Alice
print(student1.age)   # Output: 20
print(student1.grade) # Output: A

# Using the __str__ method
print(student1)  # Output: Student Name: Alice, Age: 20, Grade: A

# Updating the grade
student1.update_grade("A+")
print(student1)  # Output: Student Name: Alice, Age: 20, Grade: A+
```

# Understanding the self Parameter
**Purpose:**
* Instance Reference: self refers to the instance of the class. When a method is called on an instance, self allows the method to access and modify the attributes and other methods of the instance.

* Distinguishing Between Class and Instance Attributes: self differentiates between instance attributes and local variables. Attributes prefixed with self. belong to the instance, while those without belong to the method scope.


**How self Works:**
* Instance Creation: When an instance of Student is created, such as student1 = Student("Alice", 20, "A"), Python implicitly passes the instance (student1) to the __init__ method as the self parameter.

* Accessing Attributes: Inside the __init__ method, self.name refers to the name attribute of the instance being created (student1 in this case). Similarly, self.age and self.grade refer to the age and grade attributes.

* Calling Methods: When you call a method on an instance, such as student1.update_grade("A+"), the instance (student1) is passed as the self parameter to the update_grade method, allowing it to modify student1's attributes.
